````python
import pandas as pd
from google.colab import files
import os

file_path = '/content/vehicle_registrations_500k.csv'

# Check if the file already exists at the specified path
if not os.path.exists(file_path):
  print(f"File not found at '{file_path}'. Please upload the file.")
  uploaded = files.upload() # This will open a file selection dialog
  if uploaded:
    # Assuming the user uploaded 'vehicle_registrations_500k.csv' or a similar file
    # We take the first uploaded file's name. The user might have uploaded it with a different name.
    uploaded_filename = list(uploaded.keys())[0]
    print(f"Using uploaded file: '{uploaded_filename}'")
    df = pd.read_csv(uploaded_filename, on_bad_lines='skip')
  else:
    print("No file was uploaded. Creating an empty DataFrame.")
    df = pd.DataFrame() # Create an empty DataFrame to prevent subsequent errors
else:
  # If the file exists, proceed as normal
  df = pd.read_csv(file_path, on_bad_lines='skip')

display(df.head())
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each unique vehicle category
vehicle_category_counts = df['vehicleCategoryName'].value_counts().reset_index()
vehicle_category_counts.columns = ['vehicleCategoryName', 'count']

plt.figure(figsize=(10, 7))
sns.barplot(x='vehicleCategoryName', y='count', data=vehicle_category_counts, palette='viridis', hue='vehicleCategoryName', legend=False)
plt.title('Distribution of Vehicle Categories')
plt.xlabel('Vehicle Category')
plt.ylabel('Number of Registrations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop NaN values for accurate plotting, or choose to fill them if appropriate for the analysis.
# For outlier detection, dropping them is usually a good starting point.
weight_data = df['grossVehicleWeight'].dropna()

plt.figure(figsize=(10, 6))
sns.boxplot(x=weight_data)
plt.title('Box Plot of Gross Vehicle Weight')
plt.xlabel('Gross Vehicle Weight')
plt.show()

# Also print some descriptive statistics for context, especially quartiles and IQR
Q1 = weight_data.quantile(0.25)
Q3 = weight_data.quantile(0.75)
IQR = Q3 - Q1

print(f"\nDescriptive Statistics for Gross Vehicle Weight (excluding NaNs):\n")
print(weight_data.describe())
print(f"\nQ1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower bound for outlier detection (Q1 - 1.5 * IQR): {Q1 - 1.5 * IQR}")
print(f"Upper bound for outlier detection (Q3 + 1.5 * IQR): {Q3 + 1.5 * IQR}")
````

````python
# Convert 'registrationYear' to numeric, coercing errors to NaN
df['registrationYear'] = pd.to_numeric(df['registrationYear'], errors='coerce')

# Define a reasonable range for registration years
min_year = 1900
max_year = 2025 # Allowing for current year and potentially next year's early registrations

# Replace years outside the reasonable range with NaN
df.loc[(df['registrationYear'] < min_year) | (df['registrationYear'] > max_year), 'registrationYear'] = pd.NA

# Display the count of NaN values introduced or existing after cleaning
print(f"Number of NaN values in 'registrationYear' after cleaning: {df['registrationYear'].isnull().sum()}")

# Display the unique values again to verify cleaning
print(df['registrationYear'].unique())
````

````python
# Convert 'grossVehicleWeight' to numeric, coercing errors to NaN
df['grossVehicleWeight'] = pd.to_numeric(df['grossVehicleWeight'], errors='coerce')

# Display the count of NaN values introduced
print(f"Number of NaN values in 'grossVehicleWeight' after conversion: {df['grossVehicleWeight'].isnull().sum()}")

# Display the updated column information to confirm the data type
df.info()
````

The dataset is loaded as a DataFrame named `df`.

## Identify Electric Vehicle Fuel Types

### Subtask:
Inspect the 'fuelName' column to identify the specific string(s) used to categorize electric vehicles. This will help in filtering the dataset accurately.


**Reasoning**:
To identify specific strings for electric vehicles, I need to inspect all unique values in the 'fuelName' column.



````python
print(df['fuelName'].unique())
````

From the unique values in the 'fuelName' column, the following strings are identified as representing electric vehicles:

*   `ELECTRIC(BOV)`
*   `STRONG HYBRID EV`
*   `PURE EV`

These will be used in subsequent steps to filter the dataset for electric vehicle registrations.

## Filter for Electric Vehicles

### Subtask:
Create a new DataFrame containing only the registrations identified as electric vehicles based on the 'fuelName' column. Handle any potential missing or inconsistent data in the 'fuelName' column.


**Reasoning**:
To filter the DataFrame for electric vehicles, I need to define a list of electric vehicle fuel types and then use this list to select the relevant rows from the 'fuelName' column.



````python
ev_fuel_types = ['ELECTRIC(BOV)', 'STRONG HYBRID EV', 'PURE EV']
ev_df = df[df['fuelName'].isin(ev_fuel_types)]
display(ev_df.head())
````

## Aggregate EV Registrations by Year

### Subtask:
Group the filtered electric vehicle data by 'registrationYear' and count the number of registrations for each year to get the annual trend.


**Reasoning**:
To get the annual trend of EV registrations, I will group the 'ev_df' DataFrame by 'registrationYear' and count the occurrences for each year.



````python
ev_registrations_yearly = ev_df.groupby('registrationYear').size().reset_index(name='registrationCount')
display(ev_registrations_yearly.head())
````

## Visualize EV Registration Trend

### Subtask:
Generate a line plot to visualize the trend of electric vehicle registrations over time, showing the number of registrations per year. The plot should have appropriate labels and a title.

### Reasoning:
To visualize the trend, I will use a line plot with 'registrationYear' on the x-axis and 'registrationCount' on the y-axis from the `ev_registrations_yearly` DataFrame. This will effectively show how EV registrations have changed year over year.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='registrationYear', y='registrationCount', data=ev_registrations_yearly, marker='o')

plt.xlabel('Registration Year')
plt.ylabel('Number of EV Registrations')
plt.grid(True)
plt.xticks(rotation=45)

```

## Visualize EV Registration Trend

### Subtask:
Generate a line plot to visualize the trend of electric vehicle registrations over time, showing the number of registrations per year. The plot should have appropriate labels and a title.


**Reasoning**:
To visualize the trend of electric vehicle registrations over time, I will generate a line plot using the `ev_registrations_yearly` DataFrame, as specified in the instructions, and add appropriate labels and title.



````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='registrationYear', y='registrationCount', data=ev_registrations_yearly, marker='o')
plt.title('Electric Vehicle Registration Trend Over Time')
plt.xlabel('Registration Year')
plt.ylabel('Number of EV Registrations')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````
